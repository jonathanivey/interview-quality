```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/analysis/final_data.csv')
```

```python

predictors = ['clarity', 'immediate_relevance', 'specificity', 'attributed_meaning',
       'self_reportedness', 'spontaneity', 'n_tokens', 'token_ratio', 'rq_relevance', 'average_surprisal']
```

```python
name_mapping = {
    'clarity': 'Clarity',
    'immediate_relevance': 'Immediate\nRelevance',
    'specificity': 'Specificity',
    'attributed_meaning': 'Attributed\nMeaning',
    'self_reportedness': 'Self-Reportedness',
    'spontaneity': 'Spontaneity',
    'n_tokens': 'Response\nLength',
    'token_ratio': 'Response\nLength Ratio',
    'rq_relevance': 'RQ Relevance',
    'average_surprisal': 'Average\nSurprisal'
}

plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 10,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9
})

corr_matrix = df[predictors].rename(columns=name_mapping).corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='flare', fmt=".2f")
plt.savefig("figures/correlation_matrix.pdf", dpi=300, bbox_inches='tight')
```
