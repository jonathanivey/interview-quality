```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
analysis_df = pd.read_csv('data/analysis/final_data.csv')
```

```python
from pathlib import Path
utterance_dir = Path('data/combined_transcripts')
utterance_files = sorted(utterance_dir.glob('*.csv'))
utterance_dfs = []
for f in utterance_files:
    df = pd.read_csv(f)
    df['project_id'] = f.stem
    utterance_dfs.append(df)
utterance_df = pd.concat(utterance_dfs, ignore_index=True) if utterance_dfs else pd.DataFrame()
utterance_df['word_count'] = utterance_df['utterance'].str.split().str.len()
```

```python
from pathlib import Path
excerpt_dir = Path('data/excerpt_transcripts')
excerpt_files = sorted(excerpt_dir.glob('*.csv'))
excerpt_dfs = []
for f in excerpt_files:
    df = pd.read_csv(f)
    df['project_id'] = f.stem
    excerpt_dfs.append(df)
excerpt_df = pd.concat(excerpt_dfs, ignore_index=True) if excerpt_dfs else pd.DataFrame()
excerpt_df['word_count'] = excerpt_df['excerpt'].str.split().str.len()
```
